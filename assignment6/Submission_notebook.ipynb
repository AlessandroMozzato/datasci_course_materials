## Preparation of the data

First of all we read the files. The data is given in the CSV format. Each line is a criminal case with different attributes: time, date, location type of crime and so on. Inspecting the two files it is clear that the data for the two cities are quite different presenting very different attributes.

```python
# Importing the functions written for the assignement 6
from functions_ass6 import *
%matplotlib inline

# Reading files for Seattle and San Francisco
seattle = build_dataset('seattle')
sanfrancisco = build_dataset('sanfrancisco')
```

We decide to focus on the times of different crimes. Im particular we want to analyse time of the day and day of the week. We also tried to investigate month but unfortunately the data spans only for 3 months and we think this is not a big enough timespan to observe differences.

Extracting the crime hour from the file is relatively easy as the data is presented in the date-hh-mm-ss format for Seattle and just in hh-mm-ss for the San Francisco.

```python
# Creating necessary dataset for hour of the day and day of the week
hours_sf(sanfrancisco)
hours_seattle(seattle)
```

Extracting the day is a bit more complicated, as far as Seattle is concerned. For this city infact there is no day of the week attribute and therefore we have to build one. This can be done using the datetime library in python which converts a date from the calendar to the day of the week.
For the data from San Francisco this is not necessary as the day of the week is already provided.

```python
# Day of the week needs to be build for Seattle but is already ready for San Francisco
day_seattle(seattle)
```

In order to compare the data between the two cities we need to normalise it. To do so we consider the frequencies of crimes. For examples for the day of the week we considere the number of crimes committed during a single day over the total of the crimes, getting a percentage. This way the total number of crimes won't affect the comparison between the two cities.

```python
# Calculating frequencies of hours of the day of crimes for the two cities
hour_labels_sf,hour_freq_sf = hour_freq(sanfrancisco)
hour_labels_seattle,hour_freq_seattle = hour_freq(seattle)
# Calculating frequencies of day of the week of crimes for the two cities
day_labels_sf,day_freq_sf = day_freq(sanfrancisco)
day_labels_seattle,day_freq_seattle = day_freq(seattle)
```

# Distribution of crimes per day of the week

First of all we consider the distribution of crimes during the days of the week to see if there is any major difference.

```python
bar_plot(day_labels_sf,day_freq_sf,day_labels_seattle,day_freq_seattle,'Day','Crimes per day of the week')
```

As we can see the distribution is quite similar however some differences are clearly visible. 

As a comment of the geneal distribution is seems quite clear that the number of crimes during the week-end (friday, saturday, sunday) is sensibly higher than that of crimes during the week. 

As far as the differences between the two cities is concerned it seems quite clear that the frequency of crimes during the Week end in San Francisco is much higher than that of Seattle and conversely the frequency of crimes during the week in Seattle is higher than the number of crimes in San Francisco during the week. 

This is even more clear is we look at the frequency anomalies, i.e. the differences between the observed frequencies and the expected average if there was no trend in the data.

```python
bar_plot_freq(day_labels_sf,day_freq_sf,day_labels_seattle,day_freq_seattle,'Day','Anomaly of frequencies for Crimes per day of the week')
```

From the previous plot we can conclude that the frequency of crimes in Seattle presents much less variability with respect to the day of the week (with the exception of Friday) than the frequency of crimes in San Francisco.

# Distribution of crime per hour of the day

We analise now the frequency of crimes in Seattle and San Francisco during the different time of the day. As previously seen we consider the frequencies of crimes in order to be able to compare the data between the two cities.

```python
bar_plot(hour_labels_sf,hour_freq_sf,hour_labels_seattle,hour_freq_seattle,'Hours','Crimes per times of day')
```

As previously noticed the data seem to have similar trends but still present important differencies.

As a general consideration crime frequency tend to peak around late afternoon and decreased during the night peaking around 4. 

There are substantial differences between the two cities. In particular the frequency of crimes commited during the afternoon (13-19) is substantially higher in San Francisco than in Seattle. On the other hand the frequency of crimes commited during the night (21-4) is sensibly higher for Seattle than for San Francisco. The following differencies are confirmed by the plot of frequency anomaly similar to the one seen before.
In this case however both the frequencies have clear trend which are very different from the mean, as could be expected for the different times of day.

```python
bar_plot_freq(hour_labels_sf,hour_freq_sf,hour_labels_seattle,hour_freq_seattle,'Hours','Anomly of frequencies of Crimes per times of day')
```
